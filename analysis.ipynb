```python
import json
from pymongo import MongoClient
import pandas as pd
from IPython.display import display

# ReadOnly
connection_string = "mongodb+srv://omnipedia:omnipedia@omnipedia.y9tx7.mongodb.net/?retryWrites=true&w=majority&appName=Omnipedia"

client = MongoClient(connection_string)
db = client['omnipedia']

def extract_requirements_with_content_type(collection_name):
    # Get the collection
    collection = db[collection_name]
    
    # Perform the query to get all documents
    documents = collection.find()
    
    # Initialize a list to store the results
    results = []
    
    # Iterate over each document
    for doc in documents:
        # Extract the title and feedback
        title = doc.get('title', '')
        feedback = doc.get('feedback', {})
        
        # Iterate over each content type in feedback
        for content_type, evaluations in feedback.items():
            for evaluation in evaluations:
                section_title = evaluation.get('title', '')
                requirement_evaluations = evaluation.get('requirement_evaluations', [])
                
                # Extract all requirement IDs
                requirement_ids = [req_eval.get('requirement_id', '') for req_eval in requirement_evaluations]
                
                # Append the result
                results.append({
                    'content_type': content_type,
                    'section_title': section_title,
                    'requirement_ids': requirement_ids
                })
    
    return results

# Extract requirements from the 'atf1_wikipedia' collection
requirements_list = extract_requirements_with_content_type('atf1_wikipedia')

# Convert the results to a DataFrame for better visualization
df_requirements = pd.DataFrame(requirements_list)

# Set display options for better readability
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 100)

# Display the DataFrame directly
display(df_requirements)
```
